```python
# Cell 1: Setup and imports
import sys
from pathlib import Path
%load_ext autoreload
%autoreload 2

# Add repo root to path
repo_root = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
sys.path.insert(0, str(repo_root))

from src.utils import log_timing

from src.transformers import convert_to_json
import json

# Cell 2: Test convert_to_json
xml_path = repo_root / "xml_input" / "YHC_CL_JO_1_1-241_Catids_DEMO_all categories.xml"
output_dir = repo_root / "data" / "test_outputs"
output_dir.mkdir(parents=True, exist_ok=True)

# Call the function
with log_timing("Conversion completed in"):
    records = convert_to_json(
        xml_path=str(xml_path),
        output_dir=str(output_dir),
        remove_empty_fields=True
    )

print(f"Converted {len(records)} records")
print(f"Sample record keys: {list(records.keys())[:5]}")

for iaid, record_data in records.items():
    output_file = output_dir / f'{iaid}.json'
    with open(output_file, 'w', encoding='utf-8') as f:
        json.dump(record_data, f, indent=4)

```
